```python
import nltk 
import gensim 
from gensim import corpora, models
from nltk.stem import SnowballStemmer
from gensim.utils import simple_preprocess
from gensim.utils import tokenize
from gensim.parsing.preprocessing import STOPWORDS as stopwords
#Nach dem Einlesen der Daten importieren wir hier eine Reihe von weiteren Bibliotheken, deren Funktionen für das weitere Vorgehen von Nutzen sind
#gensim gibt uns die nötigen Werkzeuge zum Durchführen des Preprocessings, während nltk es möglich macht, einen gegeben Text in Tokens aufzuteilen
```

```python
stemmer = SnowballStemmer('german')

def stem(text):
    return stemmer.stem(text)
#Es ist notwendig einen solchen stemmer zu benutzen, damit anstelle eines bestimmten Wortes auch nach allen Wörtern mit dem gleichen Wortstamm gesucht werden kann
```

```python
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text, min_len=4): 
        if token not in stopwords:
            result.append(stem(token))
    return result
#An dieser Stelle wird die Funktion definiert, nach der das Preprocessing ablaufen wird.
#Diese teilt einen gegebenen Text in einzelne Wörter/Tokens auf und entfernt diese, die kürzer sind als 4 Buchstaben oder in der importierten Stopwortbibliothek zu finden sind.
```

```python
abstracts = (df1870_1871['text'].tolist())
#Bevor wir mit dem preprocessing beginnen können, ist es aber notwendig, zuerst den Text der Zeitungsartikel zugänglich zu machen
#Dieser wird durch diesen Befehl vom restlichen Dataframe gelöst und in eine Liste für das preprocessing umgewandelt
```

```python
from tqdm import tqdm 
#Da die folgenden Arbeitsschritte etwas länger dauern, laden wir uns hier die tqdm Bibliothek runter, mit der wir uns einen Ladebalken anzeigen lassen können
```

```python
# preprocess - process all documents
processed_docs = []
for i in tqdm(range(0, len((df1870_1871['text'].tolist())))) :
    processed_docs.append(preprocess((df1870_1871['text'].tolist()[i])))
    
print(len(processed_docs))

print(processed_docs[0:2])
#Nach den Vorbereitungen können wir nun das preprocessing über unseren Textkorpus laufen lassen
#Mit dem print-Befehl lassen wir uns hier das Ergebnis für die ersten drei Dokumente anzeigen, um zu kontrollieren, ob es so funktioniert hat wie gewollt
```

```python
# preprocess - convert word content into dictionary
dictionary = gensim.corpora.Dictionary(processed_docs) # erstellt ein dictionary, das jedem singulären Wort eine id zuordnet. Das dictionary enthält das Vokabular V des Korpus.
print(len(dictionary)) # Größe des Vokabulars

index = 0
for key, value in dictionary.iteritems():
    print(key, value)
    index +=1
    if index > 9:
        break
#Da das Programm mit einer solchen Wolke aus Wörtern noch nicht arbeiten kann, werden in diesem Schritt alle "Tokens" mit einer ID von 0 aufwärts versehen
#Dieser Schritt ist notwendig, um im folgenden zusammen mit dem stemmer feststellen zu können, wie häufig bestimmte oder sehr ähnliche Wörter im Korpus auftauchen
#Diese Zelle schließt ebenfalls mit einem print-Befehl zum überprüfen der Érgebnisse, mit einem hinzugefügten break Befehl, damit das Programm nicht zuviel Zeit mit dem Auflisten der IDs verbringt
#Stattdessen wird hier nach der zehnten ID mit dem Anzeigen aufgehört
```

```python
# preprocess - further dimensionality reduction
dictionary.filter_extremes(no_below=10, no_above=0.5, keep_n = 10000) # Funktion erlaubt es, sehr seltene und sehr häufige Wörter auszuschließen
print(len(dictionary))

bow_corpus = [dictionary.doc2bow(doc) for doc in tqdm(processed_docs)] # konvertiert ein Dokument (Liste von Wörtern) in das bag-of-words Format = 2er-Tupel (token_id, token_count). 
print(bow_corpus[0])
#Bei diesem Schritt gibt es den bereits bestehenden Erklärungen nur wenig hinzuzufügen
```

```python
# preprocess - check word stems behind IDs from dictionary

bow_doc = bow_corpus[0]

for i in tqdm(range(len(bow_doc))):
    print(f"Key {bow_doc[i][0]} = \"{dictionary[bow_doc[i][0]]}\": occurrences={bow_doc[i][1]}")
#Nach dem vorangegangenen Arbeitsschritt werden hier einmal mehr die Ergebnisse überprüft, indem die Wortstämme aus der "bag-of-words" ausgegeben werden
#Dieser Schritt beschränkt sich dabei auf das erste Dokument des Korpus, angezeigt durch die Variable Null, da dies zur Kontrolle mehr als ausreichend ist.
```

```python
# log lda algorithm
import logging
logging.basicConfig(filename = 'lda_model.log', filemode = 'w', force=True, format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
# das gensim LDA Modul berichtet nicht automatisch über den Fortschritt des Algorithmus. Mit der logging-Bibliothek können wir log-files anlegen, die uns Auskunft über
# das Modell und den Fortschritt des Algorithmus geben.
```

```python
# run LDA
id2word = dictionary
corpus = bow_corpus

lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=id2word,
                                                num_topics=10, # Anzahl der topics
                                                random_state=100, # random state zum reproduzieren der Ergebnisse
                                                chunksize=1000, # Anzahl Dokumente, die auf ein Mal in den Arbeitsspeicher geladen werden  
                                                update_every=1, # Anzahl von chunks, die auf ein Mal bearbeitet werden. 
                                                eval_every=10, # Schätzung der log-perplexity (Maß für Modellgüte)
                                                passes=10, # Anzahl wie häufig der Alg. durch den gesamten Datensatz geht
                                                alpha='symmetric', # prior für die Dokumenten-Topic Verteilung
                                                eta ='symmetric',  #  prior für die Topic-Wort Verteilung 
                                                iterations=100, # Anzahl der Iterationen
                                                per_word_topics=True) #  Wenn auf True gesetzt berechnet das Modell auch eine Liste mit wahrscheinlichsten Topics für jedes Wort

# Um zu sehen, was gensim tut müssen wir in den log file schauen!
#Nun lassen wir über unsere so vorbereiteten Wörter das LDA-Modell drüberlaufen, welches wir ebenfalls aus der vorher importierten gensim-Bibliothek haben
#Alle gesetzten Parameter sind unverändert aus dem bereitgestellten Beispielcode übernommen worden
#Zehn wurde als eine akzeptable Anzahl an Topics erachtet, da Experimente mit höheren Zahlen keine nennenswerten Veränderungen in der Zusammensetzung der Topics produzierte,
#die für unser gewähltes Thema von Relevanz war
```

```python
logging.basicConfig(filename = 'lda_model.log', filemode = 'w', force=True, format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
for index, topic in lda_model.print_topics(-1): # zeige Wortzusammensetzung der Topics, inkl. Gewichte (d.h. Bedeutung) einzelner Wörter
    print(f"Topic: {index} \nWords: {topic}")
#Angesichts der hier dargestellten Wortzusammensetzung der einzelnen Topics und unseres gewählten Themas, dem deutsch-französischen Krieg,
#kamen wir darin überein, uns mit den Topics 0, 4 und 5 genauer zu beschäftigen, da diese den größten Themenbezug aufzuwiesen schienen
```

```python
# analyze topics - main topics for each document in the collection
def analyse_topics(ldamodel, corpus, text):
    main_topic = {}
    percentage = {}
    keywords = {}
    text_snippets = {}
    
    for i, topic_list in enumerate(ldamodel [corpus]):
        topic = topic_list[0]
        topic = sorted(topic, key = lambda x: (x[1]), reverse = True) # extrahiert den wahrscheinlichsten Topicmix für jedes Dokument und sortiert die Topics. Ergebnis ist eine Liste
        # mit Topic Id (topic_num) und Anteil des Topics am Dokument (prob_topic)
       
        for j, (topic_num, prop_topic) in enumerate(topic):
            if j == 0: # extrahiere das wahrscheinlichste Topic, d.h. das mit dem größten Anteil
                wp = ldamodel.show_topic(topic_num) # extrahiere topic keywords für das Topic mit dem größten Anteil
                topic_keywords = ", ".join([word for word, prop in wp[:5]])
                main_topic[i] = int(topic_num)
                percentage[i] = round(prop_topic, 4)
                keywords[i] = topic_keywords
                text_snippets[i] = text[i][:8]
            else:
                break
    return main_topic, percentage, keywords, text_snippets
#Zum weiteren Arbeiten war es nun nötig, den Texten ihre jeweiligen Topics zuzuordnen, um damit die für uns interessanten identifizieren zu können
#Dieser Schritt ordnet jedem der Dokumente zunächst mehrere Topics zu, um anschließend das für den Text wichtigste zu extrahieren
```

```python
main_topic, percentage, keywords, text_snippets = analyse_topics(lda_model, bow_corpus, processed_docs)
```

```python
# analyze topics - print out main topic for each document in the collection

indexes = []
rows = []
for i in range(0, 10):
    indexes.append(i)
rows.append(['ID', 'Main Topic', 'Contribution (%)', 'Keywords', 'Snippet' ])

for idx in indexes:
    rows.append([str(idx), f"{main_topic.get(idx)}",
                f"{percentage.get(idx):.4f}",
                f"{keywords.get(idx)}\n",
                f"{text_snippets.get(idx)}"])
columns = zip(*rows)
column_width = [max(len(item) for item in col) for col in columns]
for row in rows:
    print(''.join(' {:{width}}'.format(row[i], width=column_width[i]) for i in range(0, len(row))))
#Ein weiterer Schritt zur Überprüfung der Ergebnisse, indem die ersten zehn Dokumente zusammen mit ihrem "Main Topic", dessen Anteil am Text und den zugehörigen Schlüsselwörtern dargestellt werden
```
